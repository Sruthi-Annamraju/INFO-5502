```python
import numpy
import matplotlib.pyplot as plt

#Input is given by the user and pi value and approximate pi value is printed.

num=int(input("\nEnter the number of points that will be generated:"))
x=numpy.random.uniform(low=-1,high=1,size=[num,1])
y=numpy.random.uniform(low=-1,high=1,size=[num,1])
inside_boolean=x**2+y**2<1
pi_approximate=4*numpy.sum(inside_boolean)/num
print('pi value: {}'.format(numpy.pi))
print('approximate value of pi: {}'.format(pi_approximate))

# To display the scattering plot

x_inside=x[inside_boolean]
y_inside=y[inside_boolean]
plt.figure(figsize=[5,5])
plt.scatter(x,y,s=1)
plt.scatter(x_inside,y_inside,color='b', s=1)
plt.show()
```
